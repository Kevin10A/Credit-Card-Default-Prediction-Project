# Step 1 Data Preparation and Exploration 
### This step involves loading the dataset and obtaining key information such as shape, column names, data types and a statistical summary. This initial exploration provides a view of the datasets structure and content. The Data preparation involves turning the data into numerical values.

```python
# Import necessary libraries 
import pandas as pd 
import matplotlib.pyplot as plt

# Load Training and test datasets 
train_data = pd.read_csv('creditdefault_train.csv')
test_data = pd.read_csv('creditdefault_test.csv')
```

```python
# Prints the shape of the dataset, whcih includes the number of rows and columns.
# It gives an overview of the size of the dataset
print("Data Shape: ", train_data.shape)

# Prints the names of all the columns present in this dataset.
# It provides information about the features available in this dataset.
print("\nColumn Names: ", train_data.columns)

# Prints the data types of each columns in the dataset. 
# It helps to understand the type of data present in this dataset
print("\nDataset Shape: ", train_data.dtypes)

# Prints a summary of statistics for each numerical column in the test dataset, including count, mean, standard deviation, minimum, maximum
# And quartile values it provides insight into the distribution and central tendency of numerical features.
print("\nDataset Statistics Summary: ", train_data.describe())

# This line plots histograms for each numerical column in the dataset. 
# It will help in understanding the range and frquency of values in each feature.
train_data.hist(figsize=(15, 15))

# This line prints the first few rows of the dataset
train_data.head()
```

```python
# Prints the shape of the dataset, whcih includes the number of rows and columns.
# It gives an overview of the size of the dataset
print("Data Shape: ", test_data.shape)

# Prints the names of all the columns present in this dataset.
# It provides information about the features available in this dataset.
print("\nColumn Names: ", test_data.columns)

# Prints the data types of each columns in the dataset. 
# It helps to understand the type of data present in this dataset
print("\nDataset Shape: ", test_data.dtypes)

# Prints a summary of statistics for each numerical column in the test dataset, including count, mean, standard deviation, minimum, maximum
# And quartile values it provides insight into the distribution and central tendency of numerical features.
print("\nDataset Statistics Summary: ", test_data.describe())

# This line plots histograms for each numerical column in the dataset. 
# It will help in understanding the range and frquency of values in each feature.
test_data.hist(figsize=(15, 15))

# This line prints the first few rows of the dataset
test_data.head()
```

```python
# Check for missing values 
print("Missing Values in training data:")
print(train_data.isnull().sum())
```

```python
# Check for missing values 
print("Missing Values in testing data:")
print(test_data.isnull().sum())
```

```python
# Encode categorical variables 
# In these datasets X2, X3, X4 are categorical variables that need to be encoded
train_data_encoded = pd.get_dummies(train_data, columns=['X2', 'X3', 'X4'], drop_first=True)
test_data_encoded = pd.get_dummies(test_data, columns=['X2', 'X3', 'X4'], drop_first=True)

# Define target variables and features 
X_train = train_data_encoded.drop('Y', axis=1)
y_train = train_data_encoded['Y']
X_test = test_data_encoded.drop('Y', axis=1)
y_test = test_data_encoded['Y']
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, roc_auc_score

# Build Decision Tree Model 
dt_model = DecisionTreeClassifier()

# Train Model 
dt_model.fit(X_train, y_train)
```

```python
# Evaluate Model and Make Predictions
y_pred_dt = dt_model.predict(X_test)

# Calculate evaluation metrics 
accuracy_dt = accuracy_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)
recall_dt = recall_score(y_test, y_pred_dt)
roc_auc_dt = roc_auc_score(y_test, y_pred_dt)

print("\nDecision Tree Model Results:")
print(f"Accuracy: {accuracy_dt}")
print(f"Precision: {precision_dt}")
print(f"F1-Score: {f1_dt}")
print(f"Recall: {recall_dt}")
print(f"AUC-ROC: {roc_auc_dt}")
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameters to tune 
param_grid_dt = {
     'max_depth': [3, 5, 7, 9],
     'min_samples_split': [2, 5, 10],
     'min_samples_leaf': [1, 2, 4]
}

# Perform Grid Search Cross Validation 
grid_search_dt = GridSearchCV(dt_model, param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X_train, y_train)

# Get best parameters and best model
best_params = grid_search_dt.best_params_
best_dt_model = grid_search_dt.best_estimator_

print("Best Parameters:", best_params)
```

```python
#Extract hyperparameters and their corresponding mean test scores from grid search results 
hyperparams_dt= grid_search_dt.cv_results_['param_max_depth'].data
mean_test_scores_dt = grid_search_dt.cv_results_['mean_test_score']

# Plot hyperparameter vs accuracy for Random Forest 
plt.plot(hyperparams_dt, mean_test_scores_dt, marker='o')
plt.title("Decision Tree: Number of Max Depth vs Accuracy")
plt.xlabel("Max Depth")
plt.ylabel('Mean Test Accuracy')
plt.grid(True)
plt.show()
```

```python
# Get best parameters and best model 
best_params_dt = grid_search_dt.best_params_

# Create a new model with the best parameters
best_dt_model = DecisionTreeClassifier(max_depth=best_params_dt['max_depth'],
                                      min_samples_split=best_params_dt['min_samples_split'],
                                      min_samples_leaf=best_params_dt['min_samples_leaf'])

# Train the model with training data
best_dt_model.fit(X_train, y_train)

# Re-Evaluate Model and Make Predictions
y_pred_dt_best = best_dt_model.predict(X_test)

# Evaluate Model performance with the best parameters 
accuracy_best_dt = accuracy_score(y_test, y_pred_dt_best)
precision_best_dt = precision_score(y_test, y_pred_dt_best)
f1_best_dt = f1_score(y_test, y_pred_dt_best)
recall_best_dt = recall_score(y_test, y_pred_dt_best)
roc_auc_best_dt = roc_auc_score(y_test, y_pred_dt_best)

print("\nDecision Tree Model (After Tuning) Results:")
print(f"Accuracy: {accuracy_best_dt}")
print(f"Precision: {precision_best_dt}")
print(f"F1-Score: {f1_best_dt}")
print(f"Recall: {recall_best_dt}")
print(f"AUC-ROC: {roc_auc_best_dt}")
```

# Evaluation and Analysis 

### F1-Score: 
### The F1-score is the mean of precision and recall. It gives a equal weight to both precision and recall and is a good metric when there is an uneven class distribution or when false positives and false negatives

### Precision:
### Precision measures the proportion of correctly predicted positive cases among all predicted cases. A high precision shows that the model has fewer false positives, meaning that when it predicts a positive outcome, It will likely be correct.

### Recall:
### Recall, also known as sensitivity or true positive rate. It measures the proportion of correctly predicted positive cases among all actual positive cases. a high recall indicates that the model can correctly identify most positives cases, thus minimizing false negatives.
